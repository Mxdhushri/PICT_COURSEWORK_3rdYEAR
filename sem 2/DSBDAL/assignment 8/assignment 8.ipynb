```python
df["rm"]=df["rm"].fillna(df["rm"].mean())
```

```python
df.isnull().sum()
```

```python
df.describe().transpose()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
corelation_matrix=df.corr().round(2)
sns.heatmap(data=corelation_matrix,annot=True)
```

```python
x=df.drop(["medv"],axis=1)
y=df["medv"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=45)
x_train
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_train,y_predict)
r2=r2_score(y_train,y_predict)
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
rmse=np.sqrt(mean_squared_error(y_train,y_predict))
r2=r2_score(y_train,y_predict)
print("The model performance for training set")
print("--------------------------------------")
print(f"Root Mean Squared Error: {rmse}")
print(f"R2 Score: {r2}")
print("\n")
```

```python
y_test_predict=model.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_test,y_test_predict))
r2=r2_score(y_test,y_test_predict)
print("The model performance for testing set")
print("--------------------------------------")
print(f"Root Mean Squared Error: {rmse}")
print(f"R2 Score: {r2}")
print("\n")
```

```python
plt.scatter(y_test, y_test_predict)

ideal_values = np.linspace(min(y_test), max(y_test)) 
# linspace makes an internal array whose starting value is minimum of y_test and end is max of y_test
plt.plot(ideal_values, ideal_values, color='purple')

plt.xlabel("Price: in $1000's")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")
plt.show()
```
