```python
bins=[15,20,25,30,35,40,45]
labels=['15-19','20-24','25-29','30-34','35-39','40-44']

df["Age_grp"]=pd.cut(df["Age"],bins=bins, labels=labels, include_lowest=True, right=False)
df.head()
```

```python
result=df[["Age_grp","Salary"]].groupby("Age_grp").agg({"Salary":["min","max","median","skew"]})
result
```
